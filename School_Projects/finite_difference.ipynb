A modified version of code I wrote for a control theory class.

Approximates a general second order linear BVP with boundary conditions:

\begin{align*}
	&{ } a_1(x)y'' +a_2(x)y'+ a_3(x) y = f(x), \quad x \in (a,b),\\
	&{ } y(a) = \alpha, \quad y(b) = \beta.
\end{align*}

```python
def bvp(a,b,alpha,beta,a1,a2,a3,f,n):
    """Computes the solution to the BVP:
        
        a1(x)y'' + a2(x)y' + a3(x)y = f
        y(a) = α, y(b) = β
        
       using finite differences."""
    #Get the x values
    xs = np.linspace(a,b,n+1)
    h = 1/n
    #Get the derivative matrices
    C = np.eye(n-1)
    D = (np.diag(np.ones(n-2), k=1) + np.diag(-np.ones(n-2), k=-1))/(2*h)
    DD = (-2*np.eye(n-1) + np.diag(np.ones(n-2), k=1) + np.diag(np.ones(n-2), k=-1))/(h**2)
    #Multiply the derivative matrices by the a functions
    for i, x in enumerate(xs[1:-1]):
        DD[i,:] *= a1(x)
    for i, x in enumerate(xs[1:-1]):
        D[i,:] *= a2(x)
    for i, x in enumerate(xs[1:-1]):
        C[i,:] *= a3(x)
    #Get the combined matrix
    DM = DD + D + C
    #Get the values of f
    fx = f(xs[1:-1])
    fx[0] -= alpha * DM[1,0]
    fx[-1] -= beta * DM[-2,-1]
    #Compute the values of u, adding the initial conditions at the end
    u = la.inv(DM)@fx
    return np.hstack([np.array([alpha]), u, np.array([beta])])
```

```python
#Get the solution
a1 = lambda x: 0.1*np.ones_like(x)
a2 = lambda x: np.zeros_like(x)
a3 = lambda x: -4*(np.pi - x**2)
f = np.cos
n = 30
a = 0
b = np.pi/2
sol = bvp(a, b, 0, 1, a1, a2, a3, f, n)
#Plot the solution
plt.plot(np.linspace(a, b, n+1), sol)
plt.title(r"Approximation of $\epsilon y''(x)-4(\pi -x^2)y(x)=cos(x)$")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#Define the problem
eps = 0.1
a1 = lambda x: eps*np.ones_like(x)
a2 = lambda x: x
a3 = lambda x: np.zeros_like(x)
f = lambda x: -eps*(np.pi**2)*np.cos(np.pi*x) - np.pi*x*np.sin(np.pi*x)
n = 150
a = -1
b = 1
#Get the solutions
sol1 = bvp(a, b, -2, 0, a1, a2, a3, f, n)
eps = 0.01
sol2 = bvp(a, b, -2, 0, a1, a2, a3, f, n)
eps = 0.001
sol3 = bvp(a, b, -2, 0, a1, a2, a3, f, n)
#Plot the solutions
plt.plot(np.linspace(a, b, n+1), sol1, label="$\epsilon =0.1$")
plt.plot(np.linspace(a, b, n+1), sol2, label="$\epsilon =0.01$")
plt.plot(np.linspace(a, b, n+1), sol3, label="$\epsilon =0.001$")
plt.legend()
plt.title(r"Approximation of $\epsilon y'' + xy' = -\epsilon \pi^2 \cos(\pi x) - \pi x \sin(\pi x)$")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
